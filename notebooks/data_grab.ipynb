```python
import requests

top = requests.get("https://hacker-news.firebaseio.com/v0/topstories.json").json()
```

```python
stories =[]
for item in top [:len(top)]:
    stories.append(requests.get(f"https://hacker-news.firebaseio.com/v0/item/{item}.json").json())
print(stories)    
```

```python
comments_id = []

for i in stories[0:]:
    comments_id+=(i['kids'])
                       
    print(comments_id)
```

```python
comments =[]
for ids in comments_id [:len(comments_id)]:
    comments.append(requests.get(f"https://hacker-news.firebaseio.com/v0/item/{ids}.json").json())
    
print(comments)
```
